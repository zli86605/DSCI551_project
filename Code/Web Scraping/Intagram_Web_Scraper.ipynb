```python
def get_post_urls(accountname):
    # Send an HTTP GET request to the user's Instagram profile page
    url = f'https://www.instagram.com/{accountname}/'

    # navigate to the user's profile page
    driver.get(url)

    # wait for the page to load
    wait = WebDriverWait(driver, 10)
    wait.until(EC.presence_of_element_located((By.TAG_NAME, 'article')))

    # scroll down to load more posts
    for i in range(50):  # adjust the number of scrolls as needed
        driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        wait.until(EC.presence_of_element_located((By.TAG_NAME, 'article')))

    # extract the URLs of the posts
    post_links = driver.find_elements_by_xpath('//article//a')
    post_urls = [link.get_attribute('href') for link in post_links]    

    return post_urls
```

```python
def get_post_likes(post_url):
    # send a GET request to the URL and store the response in a variable
    response = requests.get(post_url)

    # create a BeautifulSoup object from the response content
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # extract like number from meta
    like_number = 0

    meta = soup.find_all('meta')

    for m in meta:
        content = m.get('content')
        if content is not None:
            if 'Likes' in content:
                return content[0:content.index('Likes') - 1]
```

# Article

```python
def get_post_article(post_id):
    # URL of the Instagram post
    post_url = f'https://www.instagram.com/p/{post_id}/'

    # Send a GET request to the URL and get the HTML content
    response = requests.get(post_url)
    html_content = response.content
    
    # Create a BeautifulSoup object from the HTML content
    soup = BeautifulSoup(html_content, 'html.parser')
    
    # Extract post article
    article = soup.find('title').text
    return article
```

```python
def get_post_comments(post_id, cookies):
    post_url = 'https://www.instagram.com/p/' + post_id + '/'
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }
    response = requests.get(post_url, headers=headers, cookies=cookies)
    shortcode = post_url.split("/")[-2]
    query_hash = "97b41c52301f77ce508f55e66d17620e"
    variables = {
        "shortcode": shortcode,
        "first": 50,
    }
    url = "https://www.instagram.com/graphql/query/?query_hash={query_hash}&variables={variables}"
    url = url.format(query_hash=query_hash, variables=json.dumps(variables))
    response = requests.get(url, headers=headers, cookies=cookies)
    data = json.loads(response.text)
    comments = []
    for edge in data["data"]["shortcode_media"]["edge_media_to_parent_comment"]["edges"]:
        comment = edge["node"]["text"]
        comments.append(comment)
    return comments
```

# Image

```python
def get_post_images(post_id):
    # define the URL of the Instagram post you want to scrape
    url=f'https://www.instagram.com/p/{post_id}/'
    
    # send a GET request to the URL and store the response in a variable
    response = requests.get(url)

    # create a BeautifulSoup object from the response content
    soup = BeautifulSoup(response.content, 'html.parser')

    # find the JSON data in the HTML source code
    if len(soup.find_all('script', type='application/ld+json')) > 0:
        json_data = soup.find_all('script', type='application/ld+json')[0].contents[0]
    else:
        return []
    
    # parse the JSON data and extract the image URL(s)
    data = json.loads(json_data)
    image_urls = [i['url'] for i in data['image']]
    return image_urls
```

```python
get_post_images(df_post['post_id'][0])
```

```python
from tqdm import tqdm
df_image_urls = pd.DataFrame(columns=['post_id', 'image_urls'])

for post_id in tqdm(df_post['post_id']):
    # get the list of image urls for the current post id
    image_urls = get_post_images(post_id)
    
    # create a temporary dataframe for the current account's post URLs
    if len(image_urls) == 0:
        continue;
    else:
        temp_df = pd.DataFrame({"post_id": [post_id] * len(image_urls),
                            "image_urls": image_urls})
    
        # append the temporary dataframe to the main dataframe
        df_image_urls = pd.concat([df_image_urls, temp_df], ignore_index=True)
```
